```r
vegano_restaurantes <- vegano %>%
  # Filtra só os pratos que têm RA, RS ou RU
  filter(str_detect(prato, regex("\\b(RA|RS|RU)\\b", ignore_case = TRUE))) %>%

  # Divide corretamente em partes (cada trecho termina com ")")
  mutate(parte = str_split(prato, "\\)")) %>%
  unnest(parte) %>%

  # Remonta a parte com ")" (pois o split removeu ele)
  mutate(parte = str_trim(ifelse(parte == "", NA, paste0(parte, ")")))) %>%
  filter(!is.na(parte)) %>%

  # Extrai restaurante e prato limpo
  mutate(
    restaurante = str_extract(parte, "\\([^()]+\\)"),
    restaurante = str_remove_all(restaurante, "[()]"),
    prato_limpo = str_trim(str_remove(parte, "\\([^()]+\\)"))
  )

```

```r
padrao_restaurantes <- padrao %>%
  # Filtra só os pratos que têm RA, RS ou RU
  filter(str_detect(prato, regex("\\b(RA|RS|RU)\\b", ignore_case = TRUE))) %>%

  # Divide corretamente em partes (cada trecho termina com ")")
  mutate(parte = str_split(prato, "\\)")) %>%
  unnest(parte) %>%

  # Remonta a parte com ")" (pois o split removeu ele)
  mutate(parte = str_trim(ifelse(parte == "", NA, paste0(parte, ")")))) %>%
  filter(!is.na(parte)) %>%

  # Extrai restaurante e prato limpo
  mutate(
    restaurante = str_extract(parte, "\\([^()]+\\)"),
    restaurante = str_remove_all(restaurante, "[()]"),
    prato_limpo = str_trim(str_remove(parte, "\\([^()]+\\)"))
  )

```

```r
banco_padrao_long <- padrao_final %>%
  pivot_longer(
    cols = c(RU, RA, RS),
    names_to = "restaurante_utilizado",
    values_to = "presente"
  ) %>%
  filter(presente == 1) %>%
  select(-presente)

banco_vegano_long <- vegano_final %>%
  pivot_longer(
    cols = c(RU, RA, RS),
    names_to = "restaurante_utilizado",
    values_to = "presente"
  ) %>%
  filter(presente == 1) %>%
  select(-presente)
```
