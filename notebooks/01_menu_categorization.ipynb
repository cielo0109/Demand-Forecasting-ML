```python
# dicionário de palavras-chave -> categoria
categorias = {
    "frango": ["frango", "sobrecoxa", "nugget", "sassami"],
    "bovino": ["boi", "bovino", "bife", "carne", "lagarto", "iscas bovinas", "acém", "patinho", "almôndega"],
    "peixe": ["peixe", "tilápia", "pescada"],
    "suino": ["suíno", "bisteca", "suína", "pernil", "linguiça toscana", "copa lombo"],
    "bovino e suino": ["bovinos e suínos", "salsicha"],
    "feijoada": ["feijoada"],
    "ovo": ["fritada", "omelete"]
}

# função para categorizar
def categoriza(prato):
    if pd.isna(prato):  # handle NaN values
        return "outros"
    
    prato_lower = str(prato).lower()  # convert to string and lowercase
    for categoria, palavras in categorias.items():
        for palavra in palavras:
            if palavra in prato_lower:
                return categoria
    return "outros"

# aplica direto na coluna do seu dataframe
dataf["categoria"] = dataf["prato"].apply(categoriza)

# visualiza o resultado
print(dataf[["prato", "categoria"]])
```
